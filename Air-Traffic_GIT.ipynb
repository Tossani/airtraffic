```python
# Como iremos trabalhar com apenas os dez primeiros registros, vamos criar um novo objeto e atribuir a ele os
# 10 primeiros registros do nosso DataSet para ficar mais fácil o aprendizado. Com isto estamos criando um subset.
```

```python
df1 = df.head(10)
```

```python
df1.head(5)
```

```python
# Não necessitamos de todas as 14 colunas então vamos criar novos 3 subsets com as colunas que desejamos. 
# Atribuiremos os seguintes nomes ao novos subsets: df_cia, df_stat e df_lands
# df_cia terá as colunas: 'Operating Airline', 'Operating Airline IATA Code', 'Landing Count'
# df_stat terá as colunas: 'Aircraft Model','Aircraft Version','Landing Count','Total Landed Weight'
# df_lands terá as colunas: 'Landing Count','Total Landed Weight'
```

```python
# Utilizaremos o primeiro conceito de fatiamento ou Slice. A síntaxe é:
# subset = dataset[['Coluna1', 'Coluna2', 'Coluna3', etc...]]
```

```python
df_cia = df1[['Operating Airline','Operating Airline IATA Code','Landing Count']]
df_stat = df1[['Aircraft Model','Aircraft Version','Landing Count']]
df_lands = df1[['Operating Airline','Total Landed Weight']]
```

```python
# Para faciliar nossa compreensão, vamos trocar as colunas que estão com índice em inglês, para português. 
# Notem que utilizaremos outro conceito de manipulação de dados, substituição de índices e colunas.
# Vamos criar 3 listas com os nomes das colunas em portugues:
# colunas_cias = ['Cia Aerea','Codigo Cia Aerea','Contagem de Pousos']
# colunas_estatistica = ['Modelo Aeronave','Versao Aeronave','Contagem de Pousos', 'Peso ao Pousar']
# colunas_pousos = ['Contagem de Pousos','Peso ao Pousar']
```

```python
colunas_cias = ['Cia Aerea','Codigo Cia Aerea','Contagem de Pousos']
colunas_estatistica = ['Modelo Aeronave','Versao Aeronave','Contagem de Pousos']
colunas_pousos = ['Cia Aerea','Peso ao Pousar']
```

```python
# Agora vamos substituir as colunas do DataSet df pelas novas colunas. Vamos utilizar o atributo ".columns"
# objetonovo.columns = novacoluna
```

```python
df_cia.columns = colunas_cias
df_stat.columns = colunas_estatistica
df_lands.columns = colunas_pousos
```

```python
# Nossos subsets estão prontos para extrairmos as informações do início do Notebook.
# a) Quais as 100 maiores Cias aéreas que pousam no aeroporto?
# Vamos utilizar o subste def_cia
```

```python
# Podemos notar que o Python atribuiu um índice automático para as linhas do nosso DataFrame.
# Vamos alterar este índice e indexar o DataFrame pela coluna Operadora. Esta operação também é chamada de pivotamento.
# Para tal, vamos criar um novo objeto cia_indexada e atribuir a operação de indexação a ele. 
# Vamos utilizar o atrituto .set_index() - parâmetro: coluna que será o indexador do DataFrame: Cia Aerea.
```

```python
cia_indexada = df_cia.set_index('Cia Aerea')
cia_indexada
```

```python
# Podemos indexar um DataFrame por mais de uma coluna, basta acrescentar a coluna utilizando a notação de índice.
```

```python
cia_indexada = df_cia.set_index(['Cia Aerea', 'Codigo Cia Aerea'])
#cia_indexada
```

```python
# Segunda operação de manipulação. Vamos agrupar os dados do subset também pela Operadora. Podemos verificar que a Operadora
# ATA Airlines aparece em três índices diferentes. No DataFrame original, a ATA Airlines opera com 3 aeronaves diferentes.
# Esta é a razão pela qual ela se repete. Como queremos saber, quantos pousos esta empresa realizou, vamos agrupar os dados
# por operadora.
# Vamos criar o objeto cia_agrupada e utilizar o atributo .groupby() - parâmetro: coluna 'Cia Aerea'
```

```python
cia_agrupada = cia_indexada.groupby('Cia Aerea')
```

```python
# Podemos agrupar um DataFrame por mais de uma coluna, basta acrescentar a coluna utilizando a notação de índice.
```

```python
cia_agrupada = cia_indexada.groupby(['Cia Aerea', 'Codigo Cia Aerea'])
```

```python
# Após as operações de agrupamento e indexação podemos sumarizar nossos dados. Por exemplo: quantas ocorrências de pouso,
# temos para cada Cia Aérea? Qual o total de aeronaves que pousaram para cada Cia Aérea. 
# Para somarmos a quantidade de ocorrências de pouso para cada Cia, utilizamos o método .size()
# Para obtermos o total de pouso para cada Cia, utilizamos o método .sum()
```

```python
cia_agrupada.size()
```

```python
cia_agrupada.sum()
```

```python
# A resposa que procuramos é: Quais as Cias que mais pousaram aeronaves em nosso Dataset? Para esta resposta utilizamos
# o método .sort_values() repassando como argumento a coluna a ser ordenada, nosso caso: 'Contagem de Pousos'
```

```python
cia_agrupada.sum().sort_values(['Contagem de Pousos'])
```

```python
# Como queremos da maior para a menor, ou seja, ondem decrecente, acrescentamos no argumento da função o 'ascending=False'
cia_agrupada.sum().sort_values(['Contagem de Pousos'], ascending=False)
```

```python
# Qual tipo de aeronave que mais pousa no aeroporto?
# Vamos utilizar o DataFrame df_stat
# Porém, aqui vamos ser mais práticos, e não fazer etapa por etapa. Vamos ver o quanto o Phyton é prático,
# escrevendo todas as etapadas em uma única linha.
```

```python
df_stat_ordenado = df_stat.set_index(['Modelo Aeronave','Versao Aeronave']).groupby(['Modelo Aeronave', 'Versao Aeronave']).sum().sort_values(['Contagem de Pousos'], ascending=False)
```

```python
# Com isto, respondemos ao segundo questionamento. Qual aeronave mais utiliza a pista de pouso. Um brasileiro aparece na lista.
df_stat_ordenado
```

```python
# Vamos ao último questionamento, quais as Cias aéreas com maior quantidade de peso ao pousar? Vamos utilizar o objeto
# df_lands. Também utilizaremos uma única linha para escrever todo o código.
```

```python
df_lands
```

```python
df_lands_ordenado = df_lands.set_index('Cia Aerea').groupby('Cia Aerea').sum().sort_values(['Peso ao Pousar'], ascending=False)
```
